```python
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
columns = ["alpha", "AUC", "AP", "HR"]
df = pd.read_csv("/Users/erfanehmahmoudzadeh/Desktop/data/cora.csv", usecols=columns)
print("Contents in csv file: \n", df)
df.set_index('alpha').plot()
plt.show()
```
